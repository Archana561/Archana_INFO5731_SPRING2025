```python
#Question 2
#(a) Generate 12 random numbers between 50 and 150
nums = []  # Create an empty list
import random
for i in range(12):
    num= random.randint(50, 150)
    nums.append(num)  # Adding the number to the list
print("Generating a Random Numbers:", nums)

# (b) Sort numbers in ascending order
asc_nums = sorted(nums)
print("Asc Ord:", asc_nums)

# (c) Sort numbers in descending order
desc_nums = sorted(nums, reverse=True)
print("Desc Ord:", desc_nums)

# (d) Calculate and print the average of all numbers
avg_all = sum(nums) / len(nums)
print("Avg of all nums:", round(avg_all, 2))

# (e) Calculate and print the average of the first 6 numbers in both sorted lists
avg_f6_asc = sum(asc_nums[:6]) / 6
avg_f6_desc = sum(desc_nums[:6]) / 6

print("Avg of first 6 nums (Asc Order):", round(avg_f6_asc, 2))
print("Avg of first 6 numbers (Desc Order):", round(avg_f6_desc, 2))
```

```python
#Question 3
import string

# (a) Prompting the user to input two sentences
s1 = input("Enter the first sentence: ")
s2 = input("Enter the second sentence: ")

# (b) Converting each sentence into a list of words
s1_list = s1.split()  # Split first sentence into words
s2_list = s2.split()  # Split second sentence into words

# Combining both lists and display
all_words = s1_list + s2_list
print("Words from both sentences as a list:", all_words)

# (c) Finding common words in both sentences
common_words = list(set(s1_list) & set(s2_list))
print("Common words in both sentences:", common_words)

# (d) Removing punctuation from sentences
def remove_punctuation(sentence):
    clean_sentence = ""
    for char in sentence:
        if char not in string.punctuation:
            clean_sentence += char
    return clean_sentence

sanitized_s1 = remove_punctuation(s1)
sanitized_s2 = remove_punctuation(s2)

print("\nSanitized Sentence 1:", sanitized_s1)
print("Sanitized Sentence 2:", sanitized_s2)

# (e) Counting vowels in each sentence
def count_vowels(sentence):
    vowels = "aeiou"
    sentence = sentence.lower()
    vowel_count = {v: 0 for v in vowels}

    for char in sentence:
        if char in vowels:
            vowel_count[char] += 1

    return vowel_count

# Get vowel counts for both sentences
vowel_count1 = count_vowels(sanitized_s1)
vowel_count2 = count_vowels(sanitized_s2)

# Display vowel counts
print("\nVowel counts in Sentence 1:", vowel_count1)
print("Vowel counts in Sentence 2:", vowel_count2)

```

```python
#Question 4
import math

def calculate(expr):
    try:
        return eval(expr, {"__builtins__": None}, {
            "pow": math.pow,
            "sqrt": math.sqrt,
            "mod": lambda x, y: x % y
        })
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
    except:
        return "Error: Invalid input."

def calculator():
    print("Simple Calculator\nType 'exit' to quit.")

    while True:
        expr = input("\nEnter any expressions: ").strip()
        if expr.lower() == 'exit':
          print("Exit the calculator")
          break
        print("r:", calculate(expr))

calculator()
```

```python
#Question 5
import pandas as pd
from scipy.stats import hmean

# Loading the CSV file
df = pd.read_csv("/Assignment1_denton_housing.csv")
pivot_df = df.pivot(index="year", columns="title_field", values="value")
pivot_df = pivot_df.sort_index(ascending=True)
# (a) Calculate year-to-year difference for Occupied Housing Units
diff_ocped = pivot_df["Occupied Housing Units"].diff().dropna()
diff_ocped_sorted = diff_ocped.abs().sort_values()
# (b) Calculate percentage of vacant and occupied housing units
total_units = pivot_df["Total Housing Units"]
v_percentage = (pivot_df["Vacant Housing Units"] / total_units * 100).round(2)
ocped_percentage = (pivot_df["Occupied Housing Units"] / total_units * 100).round(2)
# (c) Find the years with the highest vacant and occupied housing units
max_v_year = pivot_df["Vacant Housing Units"].idxmax()
max_v_value = pivot_df["Vacant Housing Units"].max()
max_ocped_year = pivot_df["Occupied Housing Units"].idxmax()
max_ocped_value = pivot_df["Occupied Housing Units"].max()
# (d) Calculate the harmonic mean of total housing units
harmonic_mean_total_units = hmean(pivot_df["Total Housing Units"])
print("(a) Difference in Occupied Housing Units (Sorted in Ascending Order):")
for year, diff in diff_ocped_sorted.items():
    print(f"{int(year)}: {int(diff)}")

print("\n(b) Percentage of Vacant and Occupied Housing Units:")
print("Year    Vacant Housing Units (%)    Occupied Housing Units (%)")
for year in pivot_df.index:
    print(f"{int(year)}          {v_percentage[year]:.2f}%                                {ocped_percentage[year]:.2f}%")

print("\n(c) Years with Maximum Housing Units:")
print(f"Vacant Housing Units: {int(max_v_year)} ({max_v_value})")
print(f"Occupied Housing Units: {int(max_ocped_year)} ({max_ocped_value})")

print("\n(d) Harmonic Mean of Total Housing Units:")
print(f"{harmonic_mean_total_units:.2f}")
```

```python
#question 6
import csv
import os
import uuid

# Constants
INVENTORY_CSV_FILE = "/content/inventory.csv"

# Initialize CSV file
def initialize_inventory_data():
    if not os.path.exists(INVENTORY_CSV_FILE):
        with open(INVENTORY_CSV_FILE, mode='w', newline='') as file:
            csv_writer = csv.writer(file)
            csv_writer.writerow(["ProductCode", "ItemName", "ItemCategory", "ItemPrice", "ItemStock", "ReplenishThreshold"])

# Function to add a new product
def add_new_product():
    product_code = str(uuid.uuid4())[:8]  # Generate a unique product code
    item_name = input("Enter Product Name: ")
    item_category = input("Enter Product Category: ")
    item_price = float(input("Enter Item Price: "))
    item_stock = int(input("Enter Initial Stock Quantity: "))
    replenish_threshold = int(input("Enter Replenish Threshold Level: "))

    try:
        with open(INVENTORY_CSV_FILE, mode='a', newline='') as file:
            csv_writer = csv.writer(file)
            csv_writer.writerow([product_code, item_name, item_category, item_price, item_stock, replenish_threshold])
        print("New product added successfully!")
    except Exception as e:
        print(f"Error adding new product: {e}")

# Function to remove a product
def remove_existing_product():
    product_code = input("Enter Product Code to remove: ")
    try:
        rows_data = []
        removed = False
        with open(INVENTORY_CSV_FILE, mode='r') as file:
            csv_reader = csv.reader(file)
            for row in csv_reader:
                if row[0] != product_code:
                    rows_data.append(row)
                else:
                    removed = True
        if removed:
            with open(INVENTORY_CSV_FILE, mode='w', newline='') as file:
                csv_writer = csv.writer(file)
                csv_writer.writerows(rows_data)
            print("Product removed successfully!")
        else:
            print("Product Code not found!")
    except Exception as e:
        print(f"Error removing product: {e}")

# Function to update product details
def update_existing_product():
    product_code = input("Enter Product Code to update: ")
    try:
        rows_data = []
        updated = False
        with open(INVENTORY_CSV_FILE, mode='r') as file:
            csv_reader = csv.reader(file)
            for row in csv_reader:
                if row[0] == product_code:
                    item_name = input(f"Enter new name for {row[1]} (Leave empty to keep the same): ") or row[1]
                    item_category = input(f"Enter new category for {row[2]} (Leave empty to keep the same): ") or row[2]
                    item_price = input(f"Enter new price for {row[3]} (Leave empty to keep the same): ") or row[3]
                    item_stock = input(f"Enter new stock for {row[4]} (Leave empty to keep the same): ") or row[4]
                    replenish_threshold = input(f"Enter new replenishment level for {row[5]} (Leave empty to keep the same): ") or row[5]

                    row = [product_code, item_name, item_category, item_price, item_stock, replenish_threshold]
                    updated = True
                rows_data.append(row)

        if updated:
            with open(INVENTORY_CSV_FILE, mode='w', newline='') as file:
                csv_writer = csv.writer(file)
                csv_writer.writerows(rows_data)
            print("Product updated successfully!")
        else:
            print("Product Code not found!")
    except Exception as e:
        print(f"Error updating product: {e}")

# Function to search products
def search_for_product():
    search_term = input("Enter product name or category to search: ")
    try:
        with open(INVENTORY_CSV_FILE, mode='r') as file:
            csv_reader = csv.reader(file)
            next(csv_reader)  # Skip the header row
            search_results = [row for row in csv_reader if search_term.lower() in row[1].lower() or search_term.lower() in row[2].lower()]
            if search_results:
                for row in search_results:
                    print(f"ProductCode: {row[0]}, Name: {row[1]}, Category: {row[2]}, Price: ${row[3]}, Stock: {row[4]}")
            else:
                print("No products found!")
    except Exception as e:
        print(f"Error searching products: {e}")

# Function to restock products automatically
def auto_replenish_products():
    try:
        rows_data = []
        restocked = False
        with open(INVENTORY_CSV_FILE, mode='r') as file:
            csv_reader = csv.reader(file)
            for row in csv_reader:
                if int(row[4]) <= int(row[5]):  # If stock is less than or equal to restock level
                    row[4] = str(int(row[4]) + 20)  # Restock 20 items
                    restocked = True
                    print(f"Restocking product {row[1]} (Product Code: {row[0]})")
                rows_data.append(row)

        if restocked:
            with open(INVENTORY_CSV_FILE, mode='w', newline='') as file:
                csv_writer = csv.writer(file)
                csv_writer.writerows(rows_data)
            print("Products replenished successfully!")
        else:
            print("No products need replenishing.")
    except Exception as e:
        print(f"Error restocking products: {e}")

# Function to generate inventory value report
def generate_inventory_report():
    try:
        category_wise_report = {}
        total_inventory_value = 0

        with open(INVENTORY_CSV_FILE, mode='r') as file:
            csv_reader = csv.reader(file)
            next(csv_reader)  # Skip header row
            for row in csv_reader:
                product_value = float(row[3]) * int(row[4])
                total_inventory_value += product_value
                category_wise_report[row[2]] = category_wise_report.get(row[2], 0) + product_value

        print("\n--- Inventory Value Report ---")
        for category, value in category_wise_report.items():
            print(f"Category: {category}, Total Value: ${value:.2f}")
        print(f"\nTotal Inventory Value: ${total_inventory_value:.2f}")
    except Exception as e:
        print(f"Error generating inventory report: {e}")

# Function to check for corrupted CSV file
def verify_inventory_csv():
    try:
        with open(INVENTORY_CSV_FILE, mode='r') as file:
            csv_reader = csv.reader(file)
            header = next(csv_reader)
            if header != ["ProductCode", "ItemName", "ItemCategory", "ItemPrice", "ItemStock", "ReplenishThreshold"]:
                print("Corrupted CSV file! Restoring the default file structure.")
                initialize_inventory_data()
                print("File structure restored.")
            else:
                print("CSV file is valid.")
    except Exception as e:
        print(f"Error verifying file: {e}")

# Initialize inventory data
initialize_inventory_data()

# Menu-driven interface
while True:
    print("\nProduct Inventory Management System")
    print("1. Adding New Product")
    print("2. Removing Existing Product")
    print("3. Updating Product Details")
    print("4. Searching for Product")
    print("5. Auto Replenishing Products")
    print("6. Generating Inventory Report")
    print("7. Verifying Inventory CSV")
    print("10. Exit")
    user_choice = input("Enter your choice: ")
    if user_choice == "1":
        add_new_product()
    elif user_choice == "2":
        remove_existing_product()
    elif user_choice == "3":
        update_existing_product()
    elif user_choice == "4":
        search_for_product()
    elif user_choice == "5":
        auto_replenish_products()
    elif user_choice == "6":
        generate_inventory_report()
    elif user_choice == "7":
        verify_inventory_csv()
    elif user_choice == "10":
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
#question 7
import csv
import os
import uuid

# Constants
FLIGHT_RECORD_FILE = "/content/flights.csv"
BOOKING_RECORD_FILE = "/content/reservations.csv"

# Initialize CSV files
def initialize_files():
    if not os.path.exists(FLIGHT_RECORD_FILE):
        with open(FLIGHT_RECORD_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "Seats", "AvailableSeats", "Price"])
    if not os.path.exists(BOOKING_RECORD_FILE):
        with open(BOOKING_RECORD_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["ReservationID", "FlightID", "PassengerName"])

# Function to add new flights
def add_flight():
    flight_id = input("Enter Flight ID: ")
    origin = input("Enter Origin: ")
    destination = input("Enter Destination: ")

    while True:
        try:
            total_seats = int(input("Enter Total Seats: "))
            if total_seats <= 0:
                print("The number of seats must be a positive number greater than zero. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid number for total seats.")

    while True:
        try:
            price_per_ticket = float(input("Enter Ticket Price: "))
            if price_per_ticket <= 0:
                print("The ticket price must be a positive value. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid price.")

    try:
        with open(FLIGHT_RECORD_FILE, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([flight_id, origin, destination, total_seats, total_seats, price_per_ticket])
        print("Flight added successfully!")
    except Exception as e:
        print(f"Error adding flight: {e}")

# Function to find flights
def find_flight():
    origin_query = input("Enter Origin: ")
    destination_query = input("Enter Destination: ")

    # Edge Case: No flights found
    try:
        with open(FLIGHT_RECORD_FILE, mode='r') as file:
            reader = csv.reader(file)
            next(reader)
            found_flights = [row for row in reader if row[1].lower() == origin_query.lower() and row[2].lower() == destination_query.lower()]
            if found_flights:
                for flight in found_flights:
                    print(f"FlightID: {flight[0]}, Available Seats: {flight[4]}, Price: ${flight[5]}")
            else:
                print("No flights found!")
    except Exception as e:
        print(f"Error finding flights: {e}")

# Function to book a seat
def book_seat():
    flight_id = input("Enter Flight ID: ")
    passenger_name = input("Enter Passenger Name: ")

    # Edge Case: Invalid passenger name
    if not passenger_name.isalpha():
        print("Please enter a valid name. Special characters or numbers are not allowed.")
        return

    reservation_id = str(uuid.uuid4())[:8]  # Generate a unique reservation ID
    try:
        flight_data = []
        seat_booked = False
        with open(FLIGHT_RECORD_FILE, mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if row[0] == flight_id:
                    if int(row[4]) > 0:
                        row[4] = str(int(row[4]) - 1)
                        seat_booked = True
                    else:
                        # Edge Case: Fully booked flight
                        print(f"Sorry, this flight ({flight_id}) is fully booked.")
                        return
                flight_data.append(row)

        if seat_booked:
            with open(FLIGHT_RECORD_FILE, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(flight_data)
            with open(BOOKING_RECORD_FILE, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([reservation_id, flight_id, passenger_name])
            print(f"Seat booked successfully! Your Reservation ID is: {reservation_id}")
        else:
            print("Flight not available or fully booked!")
    except Exception as e:
        print(f"Error booking seat: {e}")

# Function to cancel a reservation
def cancel_reservation():
    reservation_id = input("Enter Reservation ID: ")

    # Edge Case: Reservation ID not found
    try:
        booking_data = []
        flight_id = ""
        reservation_canceled = False
        with open(BOOKING_RECORD_FILE, mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if row[0] == reservation_id:
                    flight_id = row[1]
                    reservation_canceled = True
                else:
                    booking_data.append(row)

        if reservation_canceled:
            with open(BOOKING_RECORD_FILE, mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerows(booking_data)
            update_seat_count(flight_id, increase=True)
            print("Reservation canceled successfully!")
        else:
            print("Reservation ID not found!")
    except Exception as e:
        print(f"Error canceling reservation: {e}")

# Function to update seat count
def update_seat_count(flight_id, increase=False):
    try:
        flight_data = []
        with open(FLIGHT_RECORD_FILE, mode='r') as file:
            reader = csv.reader(file)
            for row in reader:
                if row[0] == flight_id:
                    row[4] = str(int(row[4]) + 1 if increase else int(row[4]) - 1)
                flight_data.append(row)
        with open(FLIGHT_RECORD_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(flight_data)
    except Exception as e:
        print(f"Error updating seat count: {e}")

# Function to generate flight performance reports
def generate_report():
    try:
        with open(FLIGHT_RECORD_FILE, mode='r') as file:
            reader = csv.reader(file)
            next(reader)
            print("\n--- Flight Performance Report ---")
            for row in reader:
                tickets_sold = int(row[3]) - int(row[4])
                revenue = tickets_sold * float(row[5])
                print(f"FlightID: {row[0]}, Tickets Sold: {tickets_sold}, Revenue: ${revenue:.2f}")
    except Exception as e:
        print(f"Error generating report: {e}")

# Initialize files
initialize_files()

# Menu-driven interface
while True:
    print("\nFlight Booking System")
    print("1. Adding Flight for passengers")
    print("2. Finding Flights for passengers")
    print("3. Booking Seats for passengers")
    print("4. Cancellling Reservation for passengers")
    print("5. Generating Report for clear understanding")
    print("6. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        add_flight()
    elif choice == "2":
        find_flight()
    elif choice == "3":
        book_seat()
    elif choice == "4":
        cancel_reservation()
    elif choice == "5":
        generate_report()
    elif choice == "6":
        break
    else:
        print("Invalid choice. Please try again.")
```
